## Задание 2.1 Изучите распределение домов от наличия вида на набережную
### Постройте график
### Сделайте выводы

```python
data = df['waterfront'].value_counts()
data.index = ['no', 'yes']
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.pie(data, autopct='%1.1f%%')

plt.title('Распределение по виду на набережную')
plt.legend(data.index);
```

Вывод:только около одного процента домов имеют вид на набережную.

## Задание 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts().reset_index()
data.head()
```

```python
plt.figure(figsize=(6, 4))

sns.barplot(x=data['index'], y=data['floors'])

plt.title('Распределение по этажам')
plt.xlabel('Этаж')
plt.ylabel('Кол-во');
```

```python
Вывод: В большинстве дома имеют один или два этажа.
```

### Задание 2.2 Изучите распределение состояния домов

```python
data = df['grade'].value_counts().reset_index()
data.head()
```

```python
plt.figure(figsize=(6, 4))

sns.barplot(x=data['index'], y=data['grade'])

plt.title('Распределение домов по состоянию')
plt.xlabel('Состояние')
plt.ylabel('Кол-во');
```

```python
Вывод: В основном дома имеет средню оценку состояния(7 и 8).
```
